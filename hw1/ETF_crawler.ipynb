```python
import requests
import pandas as pd
from pandas.tseries.offsets import *
import datetime
import collections
from collections import defaultdict
```

```python
# load Bond market ETF list
bond_df = pd.read_csv("Total Bond Market ETF List (82).csv")
```

```python
select_bond = bond_df.loc[bond_df["Inception"]<"20151201","Symbol"].reset_index(drop=True)
```

```python
start_date = "20151231"
end_date = datetime.datetime.now().strftime("%Y%m%d")


index = pd.date_range(start_date, end_date, 
                      freq=BDay(),                   
                     )

columns = list(select_bond)
df = pd.DataFrame(index=index, columns=columns)

for etf in select_bond:
    url = "https://www.moneydj.com/ETF/X/xdjbcd/Basic0003BCD.xdjbcd?etfid="+etf+"&b="+start_date+"&c="+end_date
    response = requests.get(url).text.split(" ")
    trading_date = response[0].split(",")
    format_date = [d[:4]+"-"+d[4:6]+"-"+d[6:8] for d in trading_date]
    price = response[1].split(",")
    for d, p in zip(format_date, price):
        df.loc[str(d), etf] = float(p)
```

```python
df.index.name = 'Date'
```
